```python
import pandas as pd
import numpy as np
from scipy.stats import nbinom
import openpyxl
import math
file_path = 'KZK.xlsx'
df = pd.read_excel(file_path)
```

```python
data = {'Product_Name': [], 'Product_Code': [], 'v2': [], 'q2': [], 'Second_Period_Cost': []}

index = 0
while index < len(df):
    row = df.iloc[index]
    product_name = row['Product_Name']
    product_code = row['Product_Code']
    p1, p2, r1, r2 = row[['p1', 'p2', 'r1', 'r2']]
    demand1, demand2, q1_values, q2_values = row[['D1', 'D2', 'q1', 'q2']]
    v2_values = row['v2']


    v2 = 0
    while v2 <= v2_values:
        q2 = 0
        while q2 <= q2_values:
            cost_for_sec_per = 0
            d2 = 0
            while d2 <= demand2:
                D2 = nbinom.pmf(d2, r2, p2)
                cost_for_sec_per = (max((d2 - v2 - q2), 0) * D2) * 1000 + (max((v2 + q2 - d2), 0) * D2) * 1 + cost_for_sec_per
                d2 += 1
            cost_for_sec_per += q2*10
            data['Product_Name'].append(product_name)
            data['Product_Code'].append(product_code)
            data['v2'].append(v2)
            data['q2'].append(q2)
            data['Second_Period_Cost'].append(cost_for_sec_per)

            q2 += 1
        v2 += 1
    index += 1

result_rp1_output = pd.DataFrame(data)
```

```python
result_rp1_output
```

```python
def cal_min_cost_for_sec_per(df):
    min_costs_sec_2 = df.loc[df.groupby('v2')['Second_Period_Cost'].idxmin(), ['Product_Name','v2', 'q2', 'Second_Period_Cost', ]]
    min_costs_sec_2['Product_Code'] = df['Product_Code'].iloc[0]
    # Reset index after groupby
    return min_costs_sec_2.reset_index(drop=True)

product_column = 'Product_Code'
unique_products = result_rp1_output[product_column].unique()
result_list = []

for product in unique_products:
    product_df = result_rp1_output[result_rp1_output[product_column] == product].copy()
    result = cal_min_cost_for_sec_per(product_df)
    result_list.append(result)

final_result = pd.concat(result_list)
final_result.reset_index(drop=True, inplace=True)
final_result
```

```python
# Assuming df and final_result DataFrames are already defined

data_2 = {
    'Product_Name': [],
    'Product_Code': [],
    'v1': [],
    'q1': [],
    'Total Cost': []
}

# Iterate over rows using a for loop (preferable over while loop for clarity)
for index, row in df.iterrows():
    product_name = row['Product_Name']
    product_code = row['Product_Code']
    p1, p2, r1, r2 = row[['p1', 'p2', 'r1', 'r2']]
    demand1, q1_values, v1_values = row[['D1', 'q1', 'v1']]

    # Check and handle NaN values
    v1_values = 0 if pd.isna(v1_values) else int(v1_values)
    q1_values = 0 if pd.isna(q1_values) else int(q1_values)
    demand1 = 0 if pd.isna(demand1) else int(demand1)

    # Iterate over v1 values
    for v1 in range(v1_values + 1):
        # Iterate over q1 values
        for q1 in range(q1_values + 1):
            costdp = 0
            # Iterate over demand1 values
            for d1 in range(demand1 + 1):
                D1 = nbinom.pmf(d1, r1, p1)
                v2 = max(v1 + q1 - d1, 0)

                # Correct the condition with bitwise AND operator and parenthesis
                matching_rows = final_result.loc[(final_result['Product_Code'] == product_code) & (final_result['v2'] == v2), 'Second_Period_Cost']
                if not matching_rows.empty:
                    secondcost = matching_rows.values[0]
                else:
                    secondcost = 0  # Or set to an appropriate default value

                costdp += (5 * max(v1 + q1 - d1, 0) + 100 * max(d1 - q1 - v1, 0) + secondcost) * D1
            costdp += q1 * 10

            # Store the data for each combination
            data_2['Product_Name'].append(product_name)
            data_2['Product_Code'].append(product_code)
            data_2['v1'].append(v1)
            data_2['q1'].append(q1)
            data_2['Total Cost'].append(costdp)

# Create a DataFrame from the stored data
output_df = pd.DataFrame(data_2)
```

```python
result = output_df.groupby(['Product_Code', 'v1']).apply(lambda x: x.loc[x['Total Cost'].idxmin()])[['Product_Code','Product_Name', 'v1', 'q1', 'Total Cost']]
result = result.reset_index(drop=True)  # Indeksi sÄ±fÄ±rla, v1 sÃ¼tununu dÃ¼zelt
result['d1_range'] = result['v1'] + result['q1']
```

```python
expanded_rows = []

# Her bir satÄ±r iÃ§in d1_range deÄŸerini kullanarak yeni satÄ±rlar oluÅŸturup hesaplama yapalÄ±m
for index, row in result.iterrows():
    for d1 in range(row['d1_range'] + 1):
        new_row = row.copy()
        new_row['d1'] = d1
        new_row['v2'] = row['v1'] + row['q1'] - d1
        new_row['Target_First_Period'] = row['v1'] + row['q1']

        # matching_rows DataFrame'inde ilgili satÄ±rÄ± bulalÄ±m
        v2 = new_row['v2']
        product_code = new_row['Product_Code']
        matching_rows = final_result[(final_result['v2'] == v2) & (final_result['Product_Code'] == product_code)]

        # matching_rows'den q2 deÄŸerini alÄ±p yeni satÄ±ra ekleyelim
        if not matching_rows.empty:
            q2 = matching_rows.iloc[0]['q2']  # VarsayÄ±lan olarak ilk eÅŸleÅŸen satÄ±rÄ± alÄ±yoruz
            new_row['q2'] = q2

        expanded_rows.append(new_row)

# Yeni DataFrame oluÅŸtur
expanded_df = pd.DataFrame(expanded_rows)
expanded_df
```

```python
expanded_df['Target_Second_Period'] = expanded_df['v2'] + expanded_df['q2']
expanded_df = expanded_df[['Product_Name','Product_Code', 'v1', 'q1', 'Target_First_Period', 'v2', 'q2', 'Target_Second_Period']]

expanded_df
```

```python
# 'Product_Code', 'Product_Name', 'Target_First_Period', 'Target_Second_Period' sÃ¼tunlarÄ±nÄ± seÃ§erek filtreleme
filtered_df = expanded_df[['Product_Code', 'Product_Name', 'Target_First_Period', 'Target_Second_Period']]

# 'Product_Code' sÃ¼tununa gÃ¶re gruplama
grouped_df = filtered_df.groupby('Product_Code').first().reset_index()

# Sonucu gÃ¶sterme
grouped_df
```

```python
total_target_first_period = grouped_df['Target_First_Period'].sum()
total_target_second_period = grouped_df['Target_Second_Period'].sum()
print("1. Periyotta Gereken KZK,VALE,NEW_BLACK SayÄ±sÄ±:", total_target_first_period)
print("2. Periyotta Gereken KZK,VALE,NEW_BLACK SayÄ±sÄ±:", total_target_second_period)
```
